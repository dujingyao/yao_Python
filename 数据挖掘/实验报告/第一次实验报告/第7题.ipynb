```python
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_excel("text1.xlsx", header=None)
print("原始数据:")
print(df)
```

```python
# 计算每行的和
df['Sum'] = df.sum(axis=1)
print("\n每行的和:")
print(df['Sum'])
```

```python
# 第7题 - 导出为CSV文件
csv_path = "text1_sum.csv"
df.to_csv(csv_path, index=False, encoding='utf-8-sig')
print(f"\n已导出至 {csv_path}")
```
