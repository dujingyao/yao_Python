```python
import numpy as np
```

```python
# 第3题：生成数组并做 ufunc 加法（输出 result_3）
arr1 = np.arange(12)
np.random.shuffle(arr1)
arr1 = arr1.reshape(3, 4)           # 3x4
print("arr1 (3x4):\n", arr1)

arr2 = np.random.randn(16).reshape(4, 4)  # 4x4 正态分布
print("arr2 (4x4):\n", arr2)

arr1_t = arr1.T                     # 转置 -> 4x3
new_col = np.zeros((arr1_t.shape[0], 1))  # 4x1
arr1_expanded = np.hstack([arr1_t, new_col])  # 4x4
print("arr1 转置并扩充一列后 (4x4):\n", arr1_expanded)

result_3 = arr1_expanded + arr2
print("第3题结果 result_3 (4x4):\n", result_3)
```

```python
# 第4题：直接使用 result_3
repeated_result = np.repeat(result_3, 2, axis=1)  # 按列重复 2 次 -> (4,8)
print("按列重复后 shape:", repeated_result.shape)
print(repeated_result)

flattened = repeated_result.flatten()
unique_values = np.unique(flattened)
print("拉平并去重后的唯一值:\n", unique_values)
```
